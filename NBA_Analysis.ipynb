```python
import requests 
from bs4 import BeautifulSoup
data = []

year = 2005
beginning_index = 0

while year != 2025:
    response = requests.get(f'https://www.basketball-reference.com/leagues/NBA_{year}.html')
    if(response.status_code == 200):
        soup = BeautifulSoup(response.text, 'html.parser')
        for table in soup.find_all('table', id = "per_game-team"):
            rows = table.find_all('tr')
            for row in rows:
                row_data = []
                for cell in row.find_all('td'):
                    row_data.append(cell.text)
                row_data.append(year)
                data.append(row_data)
    else:
        print("Could not connect to website")
        break
    data.pop(beginning_index)
    beginning_index += 31
    year += 1 
```

```python
import pandas as pd 

headers = []
i = 0

headers = [title.text.strip() for title in table.find_all('th')]
headers = headers[1:25]
headers.append('Year')

nba_df = pd.DataFrame(data, columns = headers)
nba_df.replace(r'\*', '', regex = True)
nba_df
```

```python
league_average = nba_df[nba_df['Team'] == 'League Average']

league_average_df = pd.DataFrame(league_average, columns = headers)

nba_df = nba_df.drop(nba_df[nba_df['Team'] == 'League Average'].index)

league_average
```

```python
with pd.ExcelWriter('NBA_Stats.xlsx', engine = 'xlsxwriter') as writer:
    nba_df.to_excel(writer, sheet_name='All Team Stats', index = False)
    league_average_df.to_excel(writer, sheet_name='League Average', index = False)
```
